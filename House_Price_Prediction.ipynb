```python
df = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot']]
```

price of some house is 0, Outlier

```python
df[df['price']==0]
```

```python
df = df[df['price'] != 0]
df.shape
```

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
```

```python
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
```

```python
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]['price']
```

```python
outliers
```

```python
df = df[(df['price'] >= lower_bound) | (df['price'] <= upper_bound)]
```

```python
df.shape
```

```python
print(f'{upper_bound}, {lower_bound}')
```

```python
x = df.drop(columns='price')
```

```python
x
```

```python
y = df['price']
```

```python
print(x.shape, y.shape)
```

```python
df.describe()
```

Scaling my dataset to get better results

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print(f'Mean Square Error {mse}')
print(f'Root Mean Square Error {rmse}')
print(f'R2 Score {r2}')
```
